```python
import pandas as pd

df = pd.read_csv("titanic_cleaned.csv")

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x="Survived", data=df)

plt.show()
```

```python
sns.countplot(x="Sex", data=df)

plt.show()
```

```python
sns.countplot(

x="Sex",

hue="Survived",

data=df

)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Age"])

plt.show()
```

```python
sns.countplot(

x="Pclass",

hue="Survived",

data=df

)

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(

df.corr(numeric_only=True),

annot=True,

cmap="Blues"

)

plt.show()
```
